```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
custom_params = {"axes.spines.right": False, "axes.spines.top": False, "font.family":"EB Garamond"}
sns.set_theme(style="ticks", rc=custom_params)
```

```python
X = np.linspace(0, 2*3.14, 50)
Y = np.sin(X)
Z = np.cos(X)

sns.lineplot(x=X, y=Y)
sns.lineplot(x=X, y=Z)
```

```python
X = np.arange(1, 8)
Y = np.random.randint(1, 10, size=X.shape)
Z = np.random.randint(1, 10, size=X.shape)

sns.barplot(x=X, y=Y)
sns.lineplot(x=X-1, y=Y, color="red")
```

```python
X = np.random.random(size=1000) * np.linspace(0, 1, 1000)
Y = np.random.random(size=1000) * np.linspace(0, 1, 1000)

sns.scatterplot(x=X, y=Y)
sns.rugplot(x=X, y=Y)
```
